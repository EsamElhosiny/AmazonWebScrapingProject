```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd
import smtplib
```

```python
# Connect to the Website and get the data

url = 'https://www.amazon.eg/-/en/rango-Benzari-Rango/dp/B09GW78VP9?ref_=Oct_d_otopr_d_21864131031&pd_rd_w=m6aYh&pf_rd_p=d97d6511-ec09-4f7d-be62-83137b34bb7c&pf_rd_r=X26GGGN449FE6FTQ9PA1&pd_rd_r=ffea1570-0759-4f9a-b365-cd6ab6bc8336&pd_rd_wg=LJdxj&pd_rd_i=B09GW78VP9'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.content, 'html.parser')
title = soup.find(id="title").get_text()
price = soup.find(class_="a-price-whole").get_text()

print(title)
print(price)
```

```python
# cleaning data

title = title.strip()
price = price.strip()[:-1]

print(title)
print(price)
```

```python
# import datetime

today = datetime.date.today()
print(today)
```

```python
# create CSV file

import csv

header = ['Product', 'Price', 'Date']
data = [title, price, today]

with open('AmazonScrapingDatasets.csv', 'w', newline='', encoding= 'UTF8') as p:
    writer = csv.writer(p)
    writer.writerow(header)
    writer.writerow(data)
    
        
```
